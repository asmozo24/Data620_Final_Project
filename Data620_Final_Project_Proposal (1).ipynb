###  Project Description

#### Background

Interviews can be a nightmare for job seekers. Many companies use interviews to select the best candidate to fulfill a position.
Interview is a verbal communication with the purpose to assess the candidate's skills in relevance to the job description. 
Since it is a selection process, candidates are outto to know the job description well. Indeed, there are many studies[1,2] that have shown the effect of interview
stress on a job candidate performance. Hence, there is an opportunity for data scientist/analyst to explore solution that 
will discrease the stress among job seekers over interviews. Therefore understanding well the job description is a step to deliver the best of oneself. 


#### Research Questions

How can we use sentiment analysis to help job seekers to better understand job description?
